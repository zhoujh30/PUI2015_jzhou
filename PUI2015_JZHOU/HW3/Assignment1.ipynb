#Assignment 1 

1. GENERATE 100 samples of different sizes N (N>10 & N<2000) from each of 6 different distributions (600 samples in total), all with the same population mean. 
Include a Normal, a Poisson, a Binomial, a Chi-Squared distribution, and 2 more of your choice.                  

2. For each sample plot the sample mean against the sample size N (if you want you can do it with the sample standard deviation as well). 
Describe the behavior you see in the plots - do they look as you expected? why?

3. PLOT the distributions of all sample means (together for all distributions). 
Mandatory  - plot is as a histogram, 
Optional - plot it  in any other way you think is convincing

```python
from __future__ import print_function
__author__ = 'Jianhao Zhou'
import pylab as pl
import numpy as np
import json
s = json.load( open('fbb_matplotlibrc.json') )
pl.rcParams.update(s)
%pylab inline
```

```python
mymean = 100
md = {}
means = {}
mysize = 2000/(np.array(range(1,90))+10)
```

#Chi-Squared Distribution

```python
md['chisq']={} 
means['chisq'] = {}
allmeanschisq=[]

for n in mysize:
    md['chisq'][n] = np.random.chisquare(mymean, n)  

fig_mu_N = pl.figure(figsize=(10,10))
ax_mu_n = fig_mu_N.add_subplot(111)

for nn in md['chisq'].iterkeys():
    means['chisq'][nn] = md['chisq'][nn].mean()
    ax_mu_n.plot(nn, means['chisq'][nn], 'o')
    ax_mu_n.set_xlabel('sample size')
    ax_mu_n.set_ylabel('sample mean')
    ax_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')
    ax_mu_n.set_title('Chi-Squared Distribution')
    
for n in means['chisq']:
    allmeanschisq.append(means['chisq'][n])

pl.figure(figsize=(10,10))
pl.hist(allmeanschisq,bins=30)
pl.xlabel('sample mean')
pl.ylabel('sample size')
pl.title('Chi-Squared Distribution')
```

```python
md['normal']={} 
means['normal'] = {}
allmeansnormal=[]

for n in mysize:
    md['normal'][n] = np.random.normal(mymean, 10, n)  

fig_mu_N = pl.figure(figsize=(10,10))
ax_mu_n = fig_mu_N.add_subplot(111)

for nn in md['normal'].iterkeys():
    means['normal'][nn] = md['normal'][nn].mean()
    ax_mu_n.plot(nn, means['normal'][nn], 'o')
    ax_mu_n.set_xlabel('sample size')
    ax_mu_n.set_ylabel('sample mean')
    ax_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')
    ax_mu_n.set_title('Normal Distribution')
    
for n in means['normal']:
    allmeansnormal.append(means['normal'][n])

pl.figure(figsize=(10,10))
pl.hist(allmeansnormal,bins=30)
pl.xlabel('sample mean')
pl.ylabel('sample size')
pl.title('Normal Distribution')
```
